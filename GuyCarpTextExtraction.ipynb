```python
# imports
import pandas as pd
from openpyxl import load_workbook

import PyPDF2
import os
import io
import ace_tools
```

```python
def OpenPDF(file_name):
    '''
    OpenPDF opens a submission/pdf file and extracts the text from each page.

    Input: name of the pdf file. String
    Output: the text stored in a variable. String
    '''
    # initialize text string
    text_str = ''
    
    # Open the PDF file
    with open(file_name, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
    
        # Get the text from each page
        for page_num in range(len(reader.pages)):
            page = reader.pages[page_num] 
            text = page.extract_text()
        
            # Process the text
            # (Add your code to extract specific information here)
        
            print(text)
            text_str += text

    return text_str
```

```python
guy_text = OpenPDF("GuyCarp.pdf")
```

```python
# slice the string variable named 'guy_text' and select out only the portion after 'LOSSES' and 'ADDITIONAL INFORMATION'

# Find the index positions of 'LOSSES' and 'ADDITIONAL INFORMATION'
losses_index = guy_text.find('LOSSES')
additional_info_index = guy_text.find('ADDITIONAL INFORMATION')

# Find the index position of 'LARGE LOSSES'
large_losses_index = guy_text.find('LARGE LOSSES')

# Slice the string to get the portions after 'LOSSES' and 'ADDITIONAL INFORMATION'
losses = guy_text[losses_index + len('LOSSES'):large_losses_index] if losses_index != -1 else ''

# Slice the string to get the parts after 'LARGE LOSSES'
large_losses = guy_text[large_losses_index + len('LARGE LOSSES'):additional_info_index] if large_losses_index != -1 else ''


print("LOSSES:\n", losses)
print("\nLARGE LOSSES:\n", large_losses)
```

```python
# after RESERVE, separate data by line
reserve_index = losses.find("Reserve")
losses_lines = losses[reserve_index + len("Reserve"):]
```

```python
# separate the losses_lines by line and then space
# Convert the string data into a DataFrame

losses_columns = ['Year', '# Claims', 'Total Incurred', 'Paid', 'Reserve']

losses_df = pd.read_csv(io.StringIO(losses_lines), delim_whitespace=True, names=losses_columns)

print(losses_df)
```

```python
# Separare the Year column into 2 columns, Effective Date and Expiration Date

# Separate the 'Year' column into 'Effective Date' and 'Expiration Date'
losses_df[['Effective Date', 'Expiration Date']] = losses_df['Year'].str.split('-', expand=True)
```

```python
# Load the Excel file
excel_path = './submissions/Loss_Experience_Template.xlsx'  # Replace with the actual path to your Excel file
workbook = load_workbook(excel_path)
sheet = workbook.active
```

```python
# Define the starting row and column in the Excel sheet where the data will be inserted
start_row = 9  # Example row number where the data starts
start_col = 3  # Example column number where the data starts (D)

# Fill in the Excel sheet with the DataFrame data
for i, row in losses_df.iterrows():
    sheet.cell(row=start_row + i, column=start_col, value=row['Effective Date'])
    sheet.cell(row=start_row + i, column=start_col + 1, value=row['Expiration Date'])
    
    # TODO: extract evaluation date
    

    sheet.cell(row=start_row + i, column=start_col + 5, value=row['Total Incurred'])
    sheet.cell(row=start_row + i, column=start_col + 6, value=row['Paid'])

    # What does 'Reserve' correspond to?
    sheet.cell(row=start_row + i, column=start_col + 5, value=row['Reserve'])

# Save the updated Excel file
updated_excel_path = '/mnt/data/updated_excel_file.xlsx' 
workbook.save(updated_excel_path)

print("Excel form updated successfully!")
```
